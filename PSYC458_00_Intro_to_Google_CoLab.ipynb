# Getting started
##General Interaction
1. Cells contain code or text, this cell contains the text instructions that you are reading right now
2. To "run" a cell (i.e., execute the code/text in the cell and make something happen):
  - for text cells, hit `ctrl-enter` and it will be executed and your cursor will be moved to the next cell
    - text cells are written in markdown language, which has a simple syntax, but you can use the formatting buttons at the top to control your formatting as you like 
  - for code cells, hit `ctrl-enter` or mouse over the cell and click on the "play" icon that will appear in the top-left corner of the cell
  - `shift-enter` will execute the code/text and advance to the next cell
    - this is very convenient, since you can que the next commands this way before the previous ones have finished executing
3. Create a new cell with either the +Code or +Text buttons in the top left of this frame to create a new cell to work in!

##Integrating CoLab with google drive
We would like to be able to access data that we (eventually) upload to our google drive storage accounts, or if we would like to save processed data somewhere more permanent.
  - CoLab **does not** save your data by default, so if you do some heavy processing that you want to keep then it would be best to save it to your google drive account

1. Import some modules (`os`, which is a base python module, and `google.colab` which is specific for working within the colab environment)
2. Mount our google drive location

3. Output the current directory by `print`ing the results of the `os.getcwd()` command
  - now we know where we are in your google drive account
4. Your actual data is stored in `/content/gdrive/My Drive/`, so we need to add `My Drive` to the end of the directory string to find where our data will be
5. For convenience, colab will have created a `Colab Notebooks` directory within your google drive, you can create a subdirectory within the `Colab Notebooks` directory manually
  - right-click, create `New Folder`

6. Create a variable called `working_dir` to store the name of the directory so that you don't have to type it every time
7. Follow through with the comments and code as outlined below

```python
import os
from google.colab import drive
print(os.getcwd())
drive.mount('/content/gdrive')
print(os.getcwd())
```

```python
#create your 'data' subdirectory in google drive, then create a test file
#open it in google drive!
working_dir = '/content/gdrive/My Drive/Colab Notebooks/data/'
with open(os.path.join(working_dir,'foo.txt'), 'w') as f:
  f.write('Hello Google Drive!')

#when you need to, you can load and save other types of data to this directory (or another, if you create it!)
```

```python
#we then import numpy (which is for working with numbers, but we will use it here to load string data)
#and then load the file that we just created to see what is indside it
import numpy as np
print(np.loadtxt(os.path.join(working_dir,'foo.txt'),dtype=str))

#now go to your google drive folder 'Colab Notebooks' and open the file there as well!
```

```python
#since we have numpy here, lets also do a simple plotting example to see what data can look like!
import matplotlib.pyplot as plt
random_vector = np.random.rand(100) #100 random values between 0 and 1 (uniformly distributed)
plt.plot(random_vector)
plt.figure()
plt.plot(random_vector,'.-')

```

```python
#lets get a bit fancier and define some variables!
#here we plot the histogram of our vector to confirm that it is (approximately) uniformly distributed
num_rand = 10000
random_vector = np.random.rand(num_rand)
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(random_vector)
plt.subplot(1,2,2)
hh=plt.hist(random_vector,bins=100)

```
